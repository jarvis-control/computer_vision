```python
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/cat_image.jpg")

cv2_imshow(image)
```

```python
(b, g, r) = image[100, 50]
print("Blue:", b, "Green:", g, "Red:", r)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resized = cv2.resize(gray,(350, 300))
cv2_imshow(resized)
```

```python
cropped = gray[100:300, 150:350]
cv2_imshow(cropped)
```

```python
rotated = cv2.rotate(gray, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotated)
```

```python
flipped = cv2.flip(gray, 1)
cv2_imshow(flipped)
```

# OpenCV video streaming **codes**

```python
import cv2

cap = cv2.VideoCapture(0)

face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

while True:
    success, frame = cap.read()

    if not success:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, 1.5, 4)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    cv2.imshow("Real-time Face Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

# ==============================================

# import cv2

# cap = cv2.VideoCapture(0)

# while True:
#     success, frame = cap.read()  # True, False | Boolean

#     if not success:
#         break

#     gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
#     blurred = cv2.GaussianBlur(gray, (15, 15), 0)
#     edges1 = cv2.Canny(frame, 100, 200)
#     edges2 = cv2.Canny(frame, 150, 250)
#     flipped = cv2.flip(frame, 1)

#     cv2.imshow("Original", frame)
#     # cv2.imshow("Gray", gray)
#     # cv2.imshow("Blurred", blurred)
#     # cv2.imshow("Edges1", edges1)
#     # cv2.imshow("Edges2", edges2)
#     cv2.imshow("flipped", flipped)

#     if cv2.waitKey(1) & 0xFF == ord('q'):
#         break

# cap.release()
# cv2.destroyAllWindows()
```
